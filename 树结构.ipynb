```python
import os

def print_directory_tree(folder_path, indent=''):
    if not os.path.isdir(folder_path):
        print(f"{folder_path} 不是一个有效的文件夹路径。")
        return

    print(indent + os.path.basename(folder_path) + '/')
    indent += '|   '

    files = sorted(os.listdir(folder_path))
    for index, item in enumerate(files):
        item_path = os.path.join(folder_path, item)
        is_last = index == len(files) - 1

        if os.path.isdir(item_path):
            print_directory_tree(item_path, indent + '|   ')
        else:
            print(indent + '|-- ' + item)

        if is_last:
            print(indent[:-4] + '`-- ' + item)

# 提供要输出目录树的文件夹路径
folder_path = '/Users/jiajia/Desktop/送标数据'

print_directory_tree(folder_path)
```
